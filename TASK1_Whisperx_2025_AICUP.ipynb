## Output result

```python
# 設定音檔所在的目錄
audio_dir = "/content/audio_dataset/sample/Validation_Dataset"  # 將 "/content/audio_files" 替換為您實際的音檔目錄

# 確保輸出檔案存在
output_txt_path = "task1_answer.txt"
output_json_path = "task1_answer_timestamps.json"

# 清空輸出檔案（如果它們存在）
if os.path.exists(output_txt_path):
    os.remove(output_txt_path)
if os.path.exists(output_json_path):
    os.remove(output_json_path)

# 儲存逐字稿的字典
transcriptions = {}
timestamps = {}

def estimate_all_word_timestamps_multilingual(segment, min_duration=0.1, pause_factor=0.95):
    """
    優化版的時間戳估算函式，專為中英混合內容設計。
    核心策略：在一個沒有時間戳的區間內，根據每個單詞的「有效音節數」，按比例分配時間。
    """
    words = segment.get("words", [])
    if not words:
        return []

    # 如果所有單詞已經有時間戳，直接返回。
    # 這裡假設 WhisperX 的對齊已經給出了足夠的資訊。
    if all(word.get("start") is not None and word.get("end") is not None for word in words):
        return words

    segment_start = segment["start"]
    segment_end = segment["end"]
    new_words = []
    current_word_idx = 0

    while current_word_idx < len(words):
        word = words[current_word_idx]
        if word.get("start") is not None and word.get("end") is not None:
            new_words.append(word)
            current_word_idx += 1
            continue

        # --- 處理缺少時間戳的單詞區塊 ---
        # 區塊開始時間：如果之前有已知詞，則從其結束時間開始，否則從 segment 開始
        gap_start_time = new_words[-1]["end"] if new_words else segment_start

        # 尋找下一個已知時間戳的詞的索引
        next_known_word_idx = -1
        for i in range(current_word_idx, len(words)):
            if words[i].get("start") is not None:
                next_known_word_idx = i
                break

        # 區塊結束時間：如果找到下一個已知詞，則到其開始時間，否則到 segment 結束
        gap_end_time = words[next_known_word_idx]["start"] if next_known_word_idx != -1 else segment_end

        # 當前沒有時間戳的詞語列表
        untimestamped_words_in_gap = []
        end_of_gap_idx = next_known_word_idx if next_known_word_idx != -1 else len(words)
        for i in range(current_word_idx, end_of_gap_idx):
            untimestamped_words_in_gap.append(words[i])

        # 計算實際用於說話的時間（扣除語音間的靜默/停頓）
        # `pause_factor` 假設了一部分時間是停頓
        gap_duration = (gap_end_time - gap_start_time) * pause_factor

        # 如果這個區塊沒有無時間戳的詞，或者持續時間為零或負數，跳過
        if not untimestamped_words_in_gap or gap_duration <= 0:
            current_word_idx = end_of_gap_idx
            continue

        # 計算此區塊內所有無時間戳單詞的總音節數
        total_syllables_in_gap = sum(count_effective_syllables_v2(w["word"]) for w in untimestamped_words_in_gap)

        if total_syllables_in_gap > 0:
            # 計算每個音節的平均時間
            time_per_syllable = gap_duration / total_syllables_in_gap

            cursor = gap_start_time
            for word_in_gap in untimestamped_words_in_gap:
                word_syllables = count_effective_syllables_v2(word_in_gap["word"])
                # 預估每個詞的時長，確保不小於最小持續時間
                estimated_duration = max(word_syllables * time_per_syllable, min_duration)

                word_start = cursor
                word_end = min(word_start + estimated_duration, gap_end_time) # 確保結束時間不超過區塊結束

                new_words.append({"word": word_in_gap["word"], "start": word_start, "end": word_end})
                cursor = word_end
        else:
            # 總音節數為零時的備用方案：平均分配時間
            duration_per_word = gap_duration / len(untimestamped_words_in_gap)
            cursor = gap_start_time
            for word_in_gap in untimestamped_words_in_gap:
                word_start = cursor
                word_end = min(word_start + duration_per_word, gap_end_time)
                new_words.append({"word": word_in_gap["word"], "start": word_start, "end": word_end})
                cursor = word_end

        current_word_idx = end_of_gap_idx

    # 在所有單詞處理完畢後，確保時間戳的連續性和不超出範圍
    # 這是最後一個校準步驟，以防有微小的浮點數誤差累積
    finalized_words = []
    last_end = segment_start
    for word_info in new_words:
        current_start = max(word_info["start"], last_end) # 確保開始時間不早於前一個詞的結束時間
        current_end = max(current_start + min_duration, word_info["end"]) # 確保至少有 min_duration
        current_end = min(current_end, segment_end) # 確保不超出 segment 範圍

        finalized_words.append({"word": word_info["word"], "start": current_start, "end": current_end})
        last_end = current_end

    return finalized_words

# 處理目錄中的所有音檔
audio_files = [f for f in os.listdir(audio_dir) if f.endswith((".wav", ".mp3", ".m4a"))]
total_files = len(audio_files)

for idx, filename in enumerate(audio_files, 1):
    file_id = os.path.splitext(filename)[0]
    audio_path = os.path.join(audio_dir, filename)

    print(f"[{idx}/{total_files}] 處理檔案: {filename}")

    try:
        # 1. 使用 WhisperX 轉錄音檔
        audio = whisperx.load_audio(audio_path)
        result = model.transcribe(audio)

        # 將逐字稿添加到字典中
        full_text = " ".join([segment['text'].strip() for segment in result['segments']])
        transcriptions[file_id] = full_text

        # 2. 對齊 Whisper 輸出以獲得單詞級時間戳
        model_a, metadata = whisperx.load_align_model(language_code=result["language"], device=device)
        result = whisperx.align(result["segments"], model_a, metadata, audio, device, return_char_alignments=False)

        segments_with_timestamps = []

        for segment in result['segments']:
            if 'words' in segment:
                updated_words = estimate_all_word_timestamps_multilingual(segment) # 這裡確保是新函式名
                for word_info in updated_words:
                    word = word_info.get("word", "").strip()
                    start = word_info.get("start")
                    end = word_info.get("end")
                    if word and start is not None and end is not None:
                        segments_with_timestamps.append({
                            "word": word,
                            "start": float(start),
                            "end": float(end)
                        })

        timestamps[file_id] = {"text": full_text, "segments": segments_with_timestamps}

    except Exception as e:
        print(f"處理檔案 {filename} 時發生錯誤: {e}")

# 將逐字稿寫入 task1_answer.txt 檔案
with open(output_txt_path, "w", encoding="utf-8") as f:
    for file_id, text in transcriptions.items():
        f.write(f"{file_id}\t{text}\n")

# 將帶有時間戳的逐字稿寫入 task1_answer_timestamps.json 檔案
with open(output_json_path, "w", encoding="utf-8") as f:
    json.dump(timestamps, f, ensure_ascii=False, indent=4)

print("轉錄和對齊完成！結果已保存到 task1_answer.txt 和 task1_answer_timestamps.json")
```
